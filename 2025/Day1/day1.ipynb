```python
from functools import reduce
with open('input.txt') as f:
    lines = f.readlines()
```

```python
cum = 50
result = 0
for lin in lines:
    val = int(lin[1:])
    if lin[0] =='L': cum -= val
    else:            cum += val
    cum %= 100
    if cum == 0: result += 1
print(result)
```

```python
cum = 50
result = 0
for lin in lines:
    cum += (-1 if lin[0] =='L' else 1) * int(lin[1:])
    result += abs(cum // 100)
    cum %= 100
print(result)
```

# Single Line Solutions

```python
print("Part 1:", reduce(lambda acc, lin: (temp := (acc[0] + (1 - 2*(lin[0] == 'L')) * int(lin[1:])) % 100, acc[1] + (not temp)), lines, (50, 0))[1])
print("Part 2:", reduce(lambda acc, lin: ((temp := acc[0] + (1 - 2*(lin[0] == 'L')) * int(lin[1:])) % 100, acc[1] + abs(temp // 100)), lines, (50, 0))[1])
```
