```python
with open('input.txt') as f:
    lines = f.readlines()
    
nodes: list[str] = []
edges: list[tuple[str, str]] = []
for line in lines:
    cur, to = line.strip().split(':')
    nodes.append(cur)
    for n in to[1:].split(' '):
        edges.append((cur, n))
```

```python
from functools import cache

@cache
def get_value_of_node(node: str, start: str = 'you') -> int:
    if node == start:
        return 1
    total = 0
    for edge in (edge for edge in edges if edge[1] == node):
        total += get_value_of_node(edge[0], start)
    return total

get_value_of_node('out')
```

```python
dac_fft = get_value_of_node('fft', 'dac')
fft_dac = get_value_of_node('dac', 'fft')

assert dac_fft != 0 or fft_dac != 0, "At least one should be zero, otherwise my method wont work"

result = 0 
if dac_fft == 0:
    srv_fft = get_value_of_node('fft', 'svr')
    fft_dac = get_value_of_node('dac', 'fft')
    dac_out = get_value_of_node('out', 'dac')
    result = srv_fft * fft_dac * dac_out
else:
    srv_dac = get_value_of_node('dac', 'svr')
    dac_fft = get_value_of_node('fft', 'dac')
    fft_out = get_value_of_node('out', 'fft')
    result = srv_dac * dac_fft * fft_out

print(result)
```

```python
def generate_mermaid(edges: list[tuple[str, str]]) -> str:
    mermaid_str = "graph TD\n"
    for edge in edges:
        mermaid_str += f"    {edge[0]} --> {edge[1]}\n"
    return mermaid_str.replace('end', 'e_d')

with open('graph.mermaid', 'w') as f:
    f.write(generate_mermaid(edges))
```
