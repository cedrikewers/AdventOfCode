Sooooo this puzzle tried to make us solve one of the hardest NP-Problems: the 2D irregular packing problem. But that wasn't necessary at all. I just computed all trivial solutions (Where you just put every piece in it's own 3X3 box) and all impossible solutions (where even the most-optimal solution would use more tiles that are available). If we simply assume that the remaining cases are also solvable, we get the right answer.

```python
import numpy as np
from itertools import groupby
from typing import TypedDict

with open('input.txt') as f:
    lines = [l.strip() for l in f.readlines()]
    
*presents, trees_src = [list(group) for key, group in groupby(lines, lambda l: l == '') if not key]
presents = [np.array([[c for c in l] for l in p[1:]]) == '#' for p in presents]
class Tree(TypedDict):
    x: int
    y: int
    counts: np.ndarray
trees: list[Tree] =  [{'x': int(t[:t.index('x')]), 'y': int(t[t.index('x')+1:t.index(':')]), 'counts': np.array(list(map(int, t[t.index(':')+2:].split(' '))))} for t in trees_src]
```

```python
presents_num_tiles = [np.sum(p) for p in presents]

trivial = 0
impossible = 0
possibly_possible = 0
for i, tree in enumerate(trees):
    shape = (tree['y'], tree['x'])
    counts = tree['counts']
    if 9 * counts.sum() < shape[0] * shape[1]:
        trivial += 1
    elif (counts * presents_num_tiles).sum() > shape[0] * shape[1]:
        impossible += 1
    else:
        possibly_possible += 1

        
print(f'Trivial: {trivial}, Impossible: {impossible}, Maybe possible: {possibly_possible}')
print("Let's just assume everything that might be possible is possible...")
print(f'Part 1: {trivial + possibly_possible}')
```
