```python
with open('input.txt') as f:
    lines = f.readlines()
```

```python
banks = [list(map(int, line.strip())) for line in lines]
total = 0
for bank in banks:
    first_index = max(enumerate(bank[:-1]), key=lambda x: x[1])[0]
    second = max(bank[first_index+1:])
    total += bank[first_index] * 10 + second
print(total)
```

```python
def getmax(bank: list[int], index: int = 0, left: int = 11, current: int = 0) -> int:
    if left == -1:
        return current
    index += max(enumerate(bank[index:-left or None]), key=lambda x: x[1])[0]
    current += bank[index] * (10 ** left)
    return getmax(bank, index + 1, left - 1, current)
sum(getmax(bank) for bank in banks)
```
