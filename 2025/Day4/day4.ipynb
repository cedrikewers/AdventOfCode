```python
import numpy as np
with open('input.txt') as f:
    lines = f.readlines()
    
board = np.array([list(map(lambda x: x == '@', line.strip())) for line in lines])
```

```python
from itertools import product
offsets = list(product([-1, 0, 1], repeat=2))
offsets.remove((0, 0))

rolls = []
for a, b in zip(*np.where(board)):
    neighbors = 0
    for a_offset, y_offset in offsets:
        new_a,new_b = a + a_offset, b + y_offset
        if new_a < 0 or new_a >= board.shape[0] or new_b < 0 or new_b >= board.shape[1]:
            continue
        neighbors += board[new_a, new_b]
    if neighbors <= 3:
        rolls.append((a, b, neighbors))
        
print(len(rolls))
        
```

```python
from itertools import product
    
board = np.array([list(map(lambda x: x == '@', line.strip())) for line in lines])

offsets = list(product([-1, 0, 1], repeat=2))
offsets.remove((0, 0))

rolls_to_remove = []
total_removed = 0
while True:
    for a, b in zip(*np.where(board)):
        neighbors = 0
        for a_offset, y_offset in offsets:
            new_a,new_b = a + a_offset, b + y_offset
            if new_a < 0 or new_a >= board.shape[0] or new_b < 0 or new_b >= board.shape[1]:
                continue
            neighbors += board[new_a, new_b]
        if neighbors < 4:
            board[a, b] = False
            rolls_to_remove.append((a, b))
    
    if not len(rolls_to_remove):
        break
    total_removed += len(rolls_to_remove)
    rolls_to_remove = []

print(total_removed)
```
