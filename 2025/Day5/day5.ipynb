```python
with open('input.txt') as f:
    data = [l.strip() for l in f.readlines()]
    empty_line_idx = data.index('')
    ranges = [list(map(int, line.split('-'))) for line in data[:empty_line_idx]]
    ingredients = list(map(int, data[empty_line_idx + 1:]))
```

```python
fresh_cnt = 0
for i in ingredients:
    for r in ranges:
        if r[0] <= i <= r[1]:
            fresh_cnt += 1
            break
        
print(fresh_cnt)
```

```python
sorted_ranges = sorted(ranges, key=lambda x: x[0])
i = 0
while i < len(sorted_ranges) - 1:
    if sorted_ranges[i+1][0] <= sorted_ranges[i][1]:
        sorted_ranges[i][1] = max(sorted_ranges[i][1], sorted_ranges[i+1][1])
        del sorted_ranges[i+1]
    else:
        i += 1
        
sum([range[1] - range[0] + 1 for range in sorted_ranges])
```
