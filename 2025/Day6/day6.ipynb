```python
from functools import reduce
import re
with open('input-small.txt') as f:
    lines = f.readlines()
lines = [re.sub(r'\s+', ' ', line).strip().split(' ') for line in lines if line.strip()]

total = 0
for i in range(len(lines[0])):
    if lines[-1][i] == '+':
        total += sum(int(lines[j][i]) for j in range(len(lines)-1))
    else:
        total += reduce(lambda x, y: x * y, (int(lines[j][i]) for j in range(len(lines)-1)))
print(total)
```

```python
with open('input.txt') as f:
    lines = [line.strip() for line in f.readlines()]

total = 0
c = 0
while c < len(lines[0]):
    operator = lines[-1][c]
    sub_total = 0 if operator == '+' else 1
    nums = []
    column_width = max(lines[i][c:].index(' ') for i in range(len(lines)-1)) if ' ' in lines[0][c:] else len(lines[0]) - c
    for _ in range(column_width):
        cur_num = 0
        for i in range(len(lines) - 1):
            if lines[i][c] != ' ':
                cur_num *= 10
                cur_num += int(lines[i][c])
        nums.append(cur_num)
        c += 1
    total += sum(nums) if operator == '+' else reduce(lambda x, y: x * y, nums)
    c += 1
print(total)
```
