```python
import numpy as np
with open('input.txt') as f:
    lines = f.readlines()
board = np.array([[c for c in line.strip()] for line in lines])
board = board[0::2]
```

```python
splits = 0
beams = [np.where(board[0, :] == 'S')[0][0]]
for y in range(board.shape[0]):
    new_beams = set()
    for x in beams:
        if board[y, x] == '^':
            splits += 1
            new_beams.add(x-1)
            new_beams.add(x+1)
        else:
            new_beams.add(x)
    beams = list(new_beams)
print(splits)
```

```python
number_of_ways_to_get_there = np.zeros(board.shape, dtype=np.dtype('int64'))
number_of_ways_to_get_there[0, np.where(board[0, :] == 'S')[0][0]] = 1 # Assign starting point

for y in range(1,board.shape[0]):
    number_of_ways_to_get_there[y] = number_of_ways_to_get_there[y-1]
    caret_xs = np.where(board[y] == '^')[0]
    for caret_x in caret_xs:
        number_of_ways_to_get_there[y, caret_x-1] += number_of_ways_to_get_there[y-1, caret_x]
        number_of_ways_to_get_there[y, caret_x+1] += number_of_ways_to_get_there[y-1, caret_x]
    number_of_ways_to_get_there[y, caret_xs] = 0
    
print(np.sum(number_of_ways_to_get_there[-1, :]))
```
